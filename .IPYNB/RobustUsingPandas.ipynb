# Robust Scaling — Pandas
Same logic as NumPy version but works directly on a DataFrame.

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})
df
```

```python
def robust_scaler(df):
    median_vals = df.median()        # middle value per column
    q1_vals = df.quantile(0.25)      # 25th percentile
    q3_vals = df.quantile(0.75)      # 75th percentile
    iqr_vals = q3_vals - q1_vals     # spread of middle 50%
    return (df - median_vals) / iqr_vals
```

```python
scaled_df = robust_scaler(df)
print('Robust Scaled Data:')
scaled_df
```
