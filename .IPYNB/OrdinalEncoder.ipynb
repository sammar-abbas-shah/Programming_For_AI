# Ordinal Encoder
Encodes ordered categories (e.g. Poor < Average < Good) into numbers while preserving order.

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
# Keep only ordinal columns (all except first two)
ord_df = df.iloc[:, 2:]
ord_df.sample(5)
```

```python
# Features (exclude last/target column)
X = ord_df.iloc[:, :-1]
X.sample()
```

```python
# Define the order for each ordinal column explicitly
oEncoder = OrdinalEncoder(categories=[
    ['Poor', 'Average', 'Good'],
    ['School', 'UG', 'PG']
])

oEncoder.fit(X)
encoded = oEncoder.transform(X)

X_encoded = pd.DataFrame(encoded, columns=X.columns)
print(X_encoded.sample(10))
print('Categories:', oEncoder.categories_)
```
