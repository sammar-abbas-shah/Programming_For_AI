# KFold Cross-Validation — NumPy
Same as the DataFrame version but using NumPy arrays (direct index instead of `.iloc`).

```python
from sklearn.model_selection import KFold
import numpy as np
```

```python
X = np.array([[1,2],[3,4],[5,6],[7,8]])
y = np.array([1,2,3,4])
```

```python
kf = KFold(n_splits=2)

for train_index, test_index in kf.split(X):
    print('Train idx:', train_index, '| Test idx:', test_index)

    # NumPy arrays support direct index selection (no .iloc needed)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    print('X_train:', X_train, '| y_train:', y_train)
    print('X_test: ', X_test,  '| y_test: ', y_test)
    print()
```
