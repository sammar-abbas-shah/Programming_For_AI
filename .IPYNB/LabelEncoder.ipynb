# Label Encoder
Converts text labels (e.g. Yes/No) into numbers (0/1).

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
# Target column (last column)
y = df.iloc[:, -1]
y.sample()
```

```python
# Fit and transform: text → numbers
lEncoder = LabelEncoder()
encoded = lEncoder.fit_transform(y)

y_encoded = pd.DataFrame(encoded, columns=['purchased'])
print(y_encoded)
print('Classes:', lEncoder.classes_)
```
