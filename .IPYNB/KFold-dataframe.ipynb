# KFold Cross-Validation — DataFrame
Shows how KFold splits a pandas DataFrame into train/test folds.

```python
from sklearn.model_selection import KFold
import pandas as pd
```

```python
X = pd.DataFrame([[1,2],[3,4],[5,6],[7,8]])
y = pd.DataFrame([1,2,3,4])
```

```python
# n_splits=2 → 2 folds (each fold uses half the data for testing)
kf = KFold(n_splits=2)

for train_index, test_index in kf.split(X):
    print('Train idx:', train_index, '| Test idx:', test_index)

    # .iloc for label-based indexing on DataFrames
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    print('X_train:\n', X_train)
    print('X_test:\n',  X_test)
    print()
```
