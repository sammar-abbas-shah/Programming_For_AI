# Robust Scaling — NumPy
Scales data using median and IQR instead of mean/std, making it resistant to outliers.

```python
import numpy as np
```

```python
data = np.array([
    [1,  10,  100],
    [2,  20,  200],
    [3,  30,  300],
    [4,  40,  400],
    [5,  50,  500]
])
```

```python
def robust_scaler(data):
    median_vals = np.median(data, axis=0)          # middle value per column
    q1_vals = np.percentile(data, 25, axis=0)      # 25th percentile
    q3_vals = np.percentile(data, 75, axis=0)      # 75th percentile
    iqr_vals = q3_vals - q1_vals                   # spread of middle 50%
    return (data - median_vals) / iqr_vals
```

```python
scaled = robust_scaler(data)
print('Robust Scaled Data:')
print(scaled)
```
