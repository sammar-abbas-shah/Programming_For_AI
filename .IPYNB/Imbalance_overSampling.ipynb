# Handling Imbalanced Data — Oversampling
When one class (e.g. Fraud) has far fewer samples than another (Normal), models become biased. Oversampling fixes this.

```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import RandomOverSampler, SMOTE
```

```python
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
# Separate features (X) and target (y)
columns = [c for c in data.columns if c != 'Class']
X = data[columns]
y = data['Class']
print(X.shape, y.shape)
```

```python
# Check how imbalanced the classes are
fraud  = data[data['Class'] == 1]
normal = data[data['Class'] == 0]
print('Fraud: ', fraud.shape)
print('Normal:', normal.shape)
print(data['Class'].value_counts(normalize=True) * 100)
print(np.bincount(data['Class']))
```

## Method 1 — Random Over Sampler
Duplicates random minority-class samples until classes are balanced.

```python
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
print('Resampled shapes:', X_res.shape, y_res.shape)
X_res.head()
```

## Method 2 — SMOTE
Synthetically generates new minority-class samples (smarter than simple duplication).

```python
smote = SMOTE(random_state=42)
X_res1, y_res1 = smote.fit_resample(X, y)
print('Resampled shapes:', X_res1.shape, y_res1.shape)
X_res1.head()
```
