# Stratified KFold Demo
Unlike regular KFold, StratifiedKFold ensures every fold has the **same class ratio** as the full dataset — important for imbalanced data.

```python
from sklearn.model_selection import StratifiedKFold
import pandas as pd
```

```python
# 5 Pass (1) and 4 Fail (0) — slightly imbalanced
dic = {
    'PF':     [90,50,40,60,30,70,30,20,80],
    'Att':    [60,50,70,50,30,40,60,35,20],
    'Result': [1, 1, 1, 1, 0, 0, 0, 0, 0]
}
dataset = pd.DataFrame(dic)
dataset
```

```python
X = dataset.iloc[:, :2]    # features
y = dataset['Result']      # target
```

```python
# skf.split(X, y) needs y to maintain class proportions per fold
skf = StratifiedKFold(n_splits=3)

for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), start=1):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    print(f'==================== Fold {fold} ====================')
    print('X_Train:\n', X_train)
    print('y_Train:\n', y_train.values)
    print('X_Test:\n',  X_test)
    print('y_Test:\n',  y_test.values)

    # Check class balance in this fold's test set
    print('Test class counts:', y_test.value_counts().to_dict())
    print()
```
