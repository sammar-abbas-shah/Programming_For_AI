# Z-Score Normalization — NumPy
Rescales data so each column has mean=0 and std=1. Sensitive to outliers.

```python
import numpy as np
```

```python
data = np.array([
    [1,  10,  100],
    [2,  20,  200],
    [3,  30,  300],
    [4,  40,  400],
    [5,  50,  500]
])
```

```python
def zscore_normalizer(data):
    mean_vals = np.mean(data, axis=0)   # average per column
    std_vals  = np.std(data, axis=0)    # spread per column
    return (data - mean_vals) / std_vals
```

```python
normalized = zscore_normalizer(data)
print('Z-score Normalized Data:')
print(normalized)
```
