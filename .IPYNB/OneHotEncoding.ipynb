## One Hot Encoding
Using both pandas `get_dummies` and sklearn `OneHotEncoder` on the cars dataset.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('cars.csv')
print(df.sample(5))
print(df.shape)
print("======Unique values in each column======")
print(df.nunique())
```

```python
# Using pandas get_dummies — encode fuel and owner
ohe_encoded = pd.get_dummies(df, columns=['fuel', 'owner'])
print(ohe_encoded.columns)
```

```python
# Drop first column to handle multicollinearity
ohe_encoded1 = pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True)
print(ohe_encoded1.columns)
print(ohe_encoded1)
```

```python
# Using sklearn OneHotEncoder (drop first to avoid multicollinearity)
ohe = OneHotEncoder(drop='first')
ohe_encoded2 = ohe.fit_transform(df[['fuel', 'owner']]).toarray()
print(ohe_encoded2)
```

```python
# Combine with non-encoded columns
overall_data = np.hstack((df[['brand', 'km_driven']].values, ohe_encoded2))
print(overall_data.shape)
```
