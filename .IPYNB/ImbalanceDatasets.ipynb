# Imbalanced Datasets — imblearn Built-in Collection
Quick way to load ready-made imbalanced datasets for practice.

```python
from imblearn.datasets import fetch_datasets
import numpy as np
```

```python
# Downloads a collection of pre-built imbalanced datasets
datasets = fetch_datasets()
print('Available datasets:', list(datasets.keys()))
```

```python
# Load the 'ecoli' dataset as an example
ecoli = datasets['ecoli']
X, y = ecoli.data, ecoli.target
print('Shape:', X.shape)
print('Class counts:', np.bincount(y + 1)) 
```
